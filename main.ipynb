```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
survey = pd.read_csv("survey_results_public.csv")
```

```python
survey
```

```python
survey.Age
```

```python
survey.set_index("ResponseId", inplace=True)
```

```python
survey
```

```python
survey.columns
```

```python
age_language_filter = (survey["Age"] == "18-24 years old") & (survey["LanguageHaveWorkedWith"].str.contains("Assembly" , na=False))
```

```python
survey[age_language_filter]
```

```python
sorted_survey = survey.sort_values(by=['ConvertedCompYearly', 'YearsCode'], ascending=[False, True])
sorted_survey
```

```python
sorted_survey["ConvertedCompYearly"].mean()
```

```python
plt.style.use('_mpl-gallery')

# data


x_list = list(survey[age_language_filter].index)

x_2 = [str(x) for x in x_list]
print(x_2)
x = x_2

y = list(survey[age_language_filter]["ConvertedCompYearly"].values)
print(y)
# plot
fig, ax = plt.subplots(figsize=(256, 16))

ax.bar(x, y, width=1, edgecolor="white", linewidth=1)
ax.set_xticklabels(x, rotation=65, ha='right')
plt.show()
```
